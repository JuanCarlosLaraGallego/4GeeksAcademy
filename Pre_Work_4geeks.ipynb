```python
def sort_odd_even(numbers):
    odd = []           # almacenará los números impares
    even = []          # almacenará los números pares

    # Recorrer la lista
    for n in numbers:
        if n % 2 != 0:     # impar
            odd.append(n)
        else:               # par
            even.append(n)

    sorted_list = []        # lista final
    sorted_list.extend(odd) # primero los impares
    sorted_list.extend(even)# luego los pares

    return sorted_list
```

```python
import datetime

people = [
	{ "name": 'Joe',   "birth_date": datetime.datetime(1986,10,24) },
	{ "name": 'Bob',   "birth_date": datetime.datetime(1975,5,24) },
	{ "name": 'Erika', "birth_date": datetime.datetime(1989,6,12) },
	{ "name": 'Dylan', "birth_date": datetime.datetime(1999,12,14) },
	{ "name": 'Steve', "birth_date": datetime.datetime(2003,4,24) }
]

def calculate_age(date_of_birth):
    today = datetime.date.today()
    age = today.year - date_of_birth.year - (
        (today.month, today.day) < (date_of_birth.month, date_of_birth.day)
    )
    return age

def format_greeting(person):
    age = calculate_age(person["birth_date"])
    return f"Hello, my name is {person['name']} and I am {age} years old"

name_list = list(map(format_greeting, people))
print(name_list)
```

```python
def matrix_builder(n):
    matrix = []                  # lista principal (la matriz)
    for i in range(n):           # bucle para crear n filas
        row = []                 # lista temporal para cada fila
        for j in range(n):       # bucle interno para crear las columnas
            row.append(1)        # añadimos el número 1
        matrix.append(row)       # añadimos la fila completa a la matriz
    return matrix
```

```python
parking_state = [
  [1,1,1],
  [0,0,0],
  [1,1,2]
]

def get_parking_lot(parking_state):
    result = {
        "total_slots": 0,
        "available_slots": 0,
        "occupied_slots": 0
    }
    for row in parking_state:
        for cell in row:
            if cell == 1 or cell == 2:           # es una plaza
                result["total_slots"] += 1
                if cell == 2:                     # libre
                    result["available_slots"] += 1
                else:                             # ocupada
                    result["occupied_slots"] += 1
    return result
```

5.2. for con range

```python
for i in range(0, len(my_list)):
    print(i, my_list[i])

#range(inicio, fin) recorre desde inicio hasta fin-1.
```

La clave está en entender qué números genera:

range(inicio, fin_exclusivo, paso)

En nuestro caso:

range(len(mi_lista)-1, -1, -1)

Lo vamos a ver con un ejemplo concreto y luego interpretamos los tres -1.

Supón:

mi_lista = ['a', 'b', 'c', 'd', 'e']
len(mi_lista) = 5


Los índices son:

posición 0 → 'a'  -> elemento 1 = a

posición 1 → 'b'  -> elemento 2 = b

posición 2 → 'c'  -> elemento 3 = c

posición 3 → 'd'  -> elemento 4 = d

posición 4 → 'e'  -> elemento 5 = e



1️⃣ Primer -1: len(mi_lista) - 1
len(mi_lista) - 1   →   5 - 1 = 4


Ese valor es el índice del último elemento.

Así que esta parte:

range(len(mi_lista)-1, -1, -1)


se convierte en:

range(4, -1, -1)

Aquí:

4 = empezamos en el índice 4 (último elemento).

Todavía no miramos los otros -1, solo sabemos que empezamos desde el final.

2️⃣ Segundo -1: el fin_exclusivo

La forma general es:

range(inicio, fin_exclusivo, paso)


En range(4, -1, -1):

inicio = 4

fin_exclusivo = -1

paso = -1

En Python, fin_exclusivo significa “hasta antes de ese valor”.
Es decir, el bucle genera valores mientras no llegue a -1.

Con un paso negativo (-1), range(4, -1, -1) va generando:

1ª vuelta: i = 4

2ª vuelta: i = 3

3ª vuelta: i = 2

4ª vuelta: i = 1

5ª vuelta: i = 0

siguiente sería -1, pero como el fin es -1 y es excluido, ahí se para.

🧠 Idea importante:

Ponemos fin_exclusivo = -1 porque queremos que llegue hasta el 0.
Como el fin no se incluye, pedimos que se detenga “antes de -1”, es decir, en 0.

Si hubiéramos escrito:

range(4, 0, -1)


Los valores serían: 4, 3, 2, 1 → no incluiría el 0, porque el 0 sería el fin_exclusivo.

3️⃣ Tercer -1: el paso

Este último parámetro indica cómo avanza el contador:

range(4, -1, -1)


paso = -1 quiere decir:
“En cada iteración, resta 1 al valor actual”.

Por eso la secuencia es:

Empieza en 4

Luego 3 (4 - 1)

Luego 2 (3 - 1)

Luego 1 (2 - 1)

Luego 0 (1 - 1)

Si restara 1 otra vez → -1, pero ahí ya se para.

Si el paso fuera positivo (por ejemplo, +1), con range(0, 5, 1) generaría: 0, 1, 2, 3, 4.

Aquí usamos -1 porque queremos ir hacia atrás.

```python
for i in range(len(my_list)-1, -1, -1):
    print(my_list[i])

#len(mi_lista)-1 → último índice. -1 (como límite) → hace que pare en 0. tercer -1 → paso negativo (orden de atrás hacia adelante).
```

Resumen cortito de los tres -1

En:

for i in range(len(mi_lista)-1, -1, -1):


len(mi_lista)-1
👉 calcula el último índice (empezar desde el final).

el segundo -1 (como fin_exclusivo)
👉 indica que el rango termina antes de -1, es decir, se queda en 0. Así incluimos el índice 0 en el recorrido.

el tercer -1 (paso)
👉 indica que vamos restando 1 en cada paso, o sea, recorremos la lista de derecha a izquierda.

El resultado es que i toma los valores:

4, 3, 2, 1, 0


Y por tanto:

mi_lista[i]


recorre la lista del último al primero.

5.4. while: repetir hasta que cambie una condición

Ejemplo de “Cuenta atrás con LIFTOFF”:

```python
x = 20
while x >= 1:
    if x % 5 == 0:
        print(f"{x}!")
    else:
        print(x)
    x -= 1

print("LIFTOFF")
```

Patrón:

Inicializas una variable (x = 20).

while comprueba la condición (x >= 1).

Dentro, actualizas x (x -= 1) para acercarte a salir del bucle.

Tienes este código:

if x % 5 == 0:

    print(f"{x}!")

else:

    print(x)


Te explico cada pieza de aquí:

1️⃣ Los paréntesis: son de la función print

En ambas líneas:


print(f"{x}!")

print(x)


Los paréntesis no tienen nada que ver con la condición ni con el f.
Son simplemente la llamada a la función:

print(...) → “llama a la función print y pásale lo que hay dentro”.

En un caso le pasas un texto formateado (f"{x}!"), en el otro la variable x.

¿Qué es la f delante de la cadena? → f-string

Esta parte:

f"{x}!"


es una f-string (formatted string literal).

La f delante de las comillas le dice a Python:

“Dentro de esta cadena, lo que vaya entre {} lo tienes que evaluar como código y sustituirlo por su valor”.

```python
nombre = "Juan"
print(f"Hola, {nombre}")
```

```python
x = 20
print(f"El valor es {x}")
```

En tu caso:

f"{x}!"


significa:

Usa el valor de x

Conviértelo a texto

Y ponle un ! justo detrás

Si x = 20, entonces:

f"{x}!"   →   "20!"

¿Y las llaves {}?

Las llaves se usan solo dentro de f-strings para decir:

“Aquí va una expresión de Python que quiero incrustar”.

f"{x}!"        # `{x}` será reemplazado por el valor de x

f"{x} euros"   # ejemplo con texto

8.5. Caso real: matrix_builder(n) (matriz de 1s)

```python
def matrix_builder(n):        #Defines una función que recibe n (tamaño de la matriz).
    m = []                    #Matriz vacía: empezarás a construirla fila a fila.
    for _ in range(n):        #Se repetirá n veces → crearás n filas. Usamos _ porque no necesitamos el índice.
        row = []              #Nueva fila vacía
        for _ in range(n):    #Otra vez n veces → cada fila tendrá n columnas.
            row.append(1)     #Añade un 1 a la fila. ras este bucle interno, row será algo como [1, 1, 1] (si n = 3).
        m.append(row)         #Añades la fila completa a la matriz. Tras todas las vueltas del bucle externo, m será
                              #[
                                  #[1,1,1],
                                  #[1,1,1],
                                  #[1,1,1]
                              #]
    return m                  #Devuelves la matriz n×n de unos.
```

8.6. Caso real más “de negocio”: parking (get_parking_lot)

Aquí mezclas listas anidadas con contadores.

```python
parking_state = [
  [1,0,1,1,0,1],
  [2,0,1,1,0,1],
  [1,0,2,1,0,1],
  [1,0,1,1,0,1],
  [1,0,1,1,0,2],
  [1,0,1,1,0,1]
]
```

Convención:

0 → pasillo (no es plaza).

1 → plaza ocupada.

2 → plaza libre.

Queremos una función:

```python
def get_parking_lot(state):                                                       #Función que recibe state, la matriz del parking.
    total = free = occ = 0                                                        #Tres contadores:
                                                                                  #total → total de plazas (ocupadas o libres).
                                                                                  #free → plazas libres.
                                                                                  #occ → plazas ocupadas.
    for row in state:                                                             #Recorre cada fila de la matriz.
                                                                                  #Primera vuelta: row = [1,0,1,1,0,1]
                                                                                  #Segunda vuelta: row = [2,0,1,1,0,1], etc.
        for cell in row:                                                          #Recorre cada “celda” dentro de la fila. cell será 1, 0 o 2 según el caso.
            if cell in (1, 2):                                                    #Si cell es 1 o 2, significa que es una plaza (no es pasillo).(1, 2) es una tupla; x in (1, 2) significa “¿x es 1 o 2?”.
                total += 1                                                        #Contamos una plaza más (libre u ocupada).
                if cell == 2:                                                     #Si vale 2 → la plaza está libre → free += 1.
                    free += 1

                if cell == 1:                                                     #Si vale 1 → la plaza está ocupada → occ += 1.
                    occ += 1                                                      #Tras recorrer toda la matriz, tienes:
                                                                                  #total = número total de plazas.
                                                                                  #free = plazas libres.
                                                                                  #occ = plazas ocupadas.

    return {"total_slots": total, "available_slots": free, "occupied_slots": occ} #Devuelves un diccionario resumen (muy típico en código real):
```

**Listas de listas (matrices)**

Ejemplo clave de tus cuadernos:

matrix_builder(n) → matriz de 1s.

parking_state en el ejercicio del parking.

```python
parking_state = [
  [1,0,1,1,0,1],
  [2,0,1,1,0,1],
  [1,0,2,1,0,1],
  ...
]
```

*Función que recorre esa matriz:*

```python
def get_parking_lot(state):
    total = free = occ = 0
    for row in state:          # primera dimensión (filas)
        for cell in row:       # segunda dimensión (columnas)
            if cell in (1, 2):
                total += 1
                if cell == 2:
                    free += 1
                if cell == 1:
                    occ += 1
    return {
        "total_slots": total,
        "available_slots": free,
        "occupied_slots": occ
    }

print(get_parking_lot(parking_state))
```

*Esta función mezcla listas anidadas, diccionarios, bucles anidados y contadores: un caso real muy completo*
